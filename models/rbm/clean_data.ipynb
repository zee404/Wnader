```python
# Load JSON data into a DataFrame
reviews_data_1 = pd.read_json("data/attractions/reviews/attractions_reviews_batch1.json")

# Print the DataFrame
reviews_data_1
```

```python
# Load JSON data into a DataFrame
reviews_data_2 = pd.read_json("data/attractions/reviews/attractions_reviews_batch2.json")

# Print the DataFrame
reviews_data_2
```

```python
# Combine DataFrames
reviews_data = pd.concat([reviews_data_1, reviews_data_2])
```

```python
reviews_data
```

```python
reviews_data.isna().sum()
```

```python
reviews_data = reviews_data.replace('nil', np.nan)
```

```python
reviews_data.isnull().sum()
```

```python
# get all unique values in the 'name' column
unique_names = reviews_data['user'].unique()

# create a new dataframe with unique names
new_df_user_name = pd.DataFrame({'user': unique_names})
new_df_user_name
```

```python
new_df_user_name["user"].is_unique
```

```python
new_df_user_name['user_id'] = new_df_user_name.index
new_df_user_name["user_id"].is_unique
```

```python
new_df_user_name.head()
```

```python
reviews_data = reviews_data.merge(new_df_user_name, on='user', how='left')
reviews_data.head()
```

```python
reviews_data.to_csv(path_or_buf = "data/attractions/reviews/attractions_reviews_clean.csv", index=False)
```
