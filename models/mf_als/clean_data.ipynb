```python
import numpy as np
import pandas as pd
```

```python
df_reviews = pd.read_csv('data/hotels/details/reviews_dedup.csv')
df_reviews.head()
```

```python
df_reviews.isnull().sum()
```

```python
df_reviews = df_reviews.replace('nil', np.nan)
```

```python
df_reviews.isnull().sum()
```

```python
df_reviews.head()
```

```python
df_reviews.dropna(subset=['user_name', 'user_profile'], how='all', inplace=True)
```

```python
df_reviews.isnull().sum()
```

```python
df_test = df_reviews[df_reviews.duplicated(subset=['user_profile'],keep=False)]
```

```python
df_test.sort_values(by=['user_profile'])
```

```python
df_test = df_test.drop(columns=['user_profile', 'user_review','user_reviewdate','user_staydate'])
```

```python
df_test.head(10) 
```

```python
df_test['user_name'] = df_test['user_name'].str.replace('A TripAdvisor Member','undefined')
```

```python
df_test = df_test.rename(columns={'user_rating': 'rating', 'id': 'hotel_id'})
```

```python
# get all unique values in the 'name' column
unique_names = df_test['user_name'].unique()

# create a new dataframe with unique names
new_df_user_name = pd.DataFrame({'user_name': unique_names})
new_df_user_name
```

```python
new_df_user_name["user_name"].is_unique
```

```python
new_df_user_name['user_id'] = new_df_user_name.index
new_df_user_name["user_id"].is_unique
```

```python

new_df_user_name.head()
```

```python
df_test = df_test.merge(new_df_user_name, on='user_name', how='left')
df_test.head()
```

```python
df_test["user_id"].is_unique
```

```python
df_test.to_csv(path_or_buf = "data/hotels/details/reviews_clean.csv", index=False)
```
